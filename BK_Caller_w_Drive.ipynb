```python
spreadsheet_id = '1_r9UF8tqLGP_5HFXEdtEs4cwCEHgMZJm25sLPqW5CUc'
sheet_id = 'Eingaben'
url = f'https://docs.google.com/spreadsheets/d/{spreadsheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_id}'

def german_number(x):
    if x == '':
        return 0
    else:
        return float( x[:-1].replace('.', '').replace(',','.') )
                                 
df = pd.read_csv(url, 
                 index_col=0,
                 usecols=range(16),
                 decimal=',',
                 converters={'Anschaffungskosten [€]': german_number,
                             'Staatliche Förderung [€]': german_number,
                             'Anzahlung [€]': german_number,
                             'Finanzierungszins [%]': german_number,
                             'Restwert [€]': german_number,
                             'Versicherungskosten [€/a]': german_number,
                             'Steuern [€/a]': german_number,
                             'Durchschnittlicher Spritpreis über Haltedauer [€/E]': german_number,
                             'Reparatur und Wartungskosten [€/a]': german_number,
                             'Leasingrate [€]': german_number})
df.tail()
```
