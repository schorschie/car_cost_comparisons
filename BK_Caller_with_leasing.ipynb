```python
latex = ''

overview = pd.DataFrame()
overview.name='Monatliche Kosten im Verlgeich'

comparison_columns = ['Preis Netto', 'Baujahr', 'Kilometerstand', 'Antrieb', 
                      'Raten', 'Versicherung', 'Steuern', 'Reparaturen', 'Verbrauch',
                      'Summe', 'Anzahlung', 'Restwert', 'Gesamtsumme']
comparison = pd.DataFrame(columns=comparison_columns)
comparison.index.name='Fahrzeug'

for car in cars:
    if car['estimate']:
        estimate_string = r'\emph{Gesch\"atzter}'
    else:
        estimate_string = r'\emph{Realer}'
        
    if car['buy']:
        label=car['label'] + ' Kredit'
        bk_inputs = {key: car[key] for key in ['acquisition_cost', 
                                               'government_funding',
                                               'down_payment',
                                               'holding_period',
                                               'millage_yearly',
                                               'consumption',
                                               'fuel_price',
                                               'repairs_yearly',  # bis hierher Hauptkosten
                                               'interest_rate',
                                               'taxes_yearly',
                                               'insurance_yearly',
                                               'rest_value']}
        (input_values, calc_values) = Operating_Cost.buy(**bk_inputs).get_frames()
    else:
        label=car['label'] + ' Leasing'
        bk_inputs = {key: car[key] for key in ['instalment', 
                                               'government_funding',
                                               'down_payment',
                                               'holding_period',
                                               'millage_yearly',
                                               'consumption',
                                               'fuel_price',
                                               'repairs_yearly',  # bis hierher Hauptkosten
                                               'taxes_yearly',
                                               'insurance_yearly']}
        (input_values, calc_values) = Operating_Cost.lease(**bk_inputs).get_frames()
        
    input_values_path = Path('images/' + slugify(label) + '-boundaries.png')
    calc_df_path = Path('images/' + slugify(label) + '-calc_df.png')
    
    comparison = comparison.append(pd.DataFrame(data=[[car['acquisition_cost'],
                                                      car['car_year'],
                                                      car['odometer'],
                                                      car['fuel_type'],
                                                      calc_values.loc['Raten', 'Monatlich'],
                                                      calc_values.loc['Versicherung', 'Monatlich'],
                                                      calc_values.loc['Steuern', 'Monatlich'],
                                                      calc_values.loc['Reparaturen', 'Monatlich'],
                                                      calc_values.loc['Verbrauch', 'Monatlich'],
                                                      calc_values.loc['Summe', 'Monatlich'],
                                                      calc_values.loc['Anzahlung', 'Monatlich'],
                                                      calc_values.loc['Restwert', 'Monatlich'],
                                                      calc_values.loc['Gesamtsumme', 'Monatlich']]],
                                                columns=comparison_columns,
                                                index=[label]))

    latex = r"""%s
\twocolumn

\section*{%s}
\begin{center}
\includegraphics[width=0.5\textheight]{%s}
\null
\vspace{0.5cm}
\includegraphics[width=0.5\textheight]{%s}
\end{center}

\pagebreak
\null
\vspace{2cm}
\begin{center}
\includegraphics[width=0.9\columnwidth]{%s}

%s
\end{center}

\begin{itemize}
    \item Baujahr: %d $\rightarrow$ Laufzeit: %.0f Jahre
    \item \href{%s}{Weblink}
    \item %s Spritpreis \"uber Haltedauer: %.2f \euro{}/%s
    \item Antrieb: %s, mit %d PS
    \item %s
\end{itemize}

\pagebreak

""" %(latex,
      label,
      input_values_path.as_posix(),
      calc_df_path.as_posix(),
      car['image_url'],
      label,
      car['car_year'],
      car['holding_period'],
      car['web_url'].replace('#', '\#'),
      estimate_string,
      car['fuel_price'],
      car['fuel_unit'],
      car['fuel_type'],
      car['power'],
      car['rest_value_thumb_value']
     )

    # calc_values.loc['Lieferzeit', :] = car['delivery_time']
    # calc_values.loc['Kraftstoffsorte', :] = car['fuel_type']
    # calc_values = calc_values.reindex(index = ['Lieferzeit',
    #                                            'Kraftstoffsorte',
    #                                            'Anschaffungskosten',
    #                                            'Zinsen',
    #                                            'Versicherung',
    #                                            'Steuern',
    #                                            'Reparaturen',
    #                                            'Verbrauch',
    #                                            'SUMME',
    #                                            'Restwert',
    #                                            'Summe mit Restwert'
    #                                           ])
    
    overview[label] = calc_values['Monatlich']
    
    input_values_path.parent.mkdir(parents=True, exist_ok=True)
    
    bold_row = pd.IndexSlice[calc_values.index[(calc_values.index == "Summe") | (calc_values.index == "Gesamtsumme")], :]
    light_row = pd.IndexSlice[calc_values.index[6:], :]
    
    dfi.export(input_values.round(2), str(input_values_path))
    dfi.export(calc_values.style.applymap(df_style, subset=bold_row)\
               .applymap(light_style, subset=light_row).set_precision(2), str(calc_df_path))
#    display(input_values.round(2))
#    display(calc_values.style.applymap(df_style, subset=bold_row)\
#               .applymap(light_style, subset=light_row).set_precision(2))
#
```
