```python
latex = ''
for car in cars:
    label=car['label']
    (input_values, calc_values) = bk(car['car_price'],
                                     car['insurance'],
                                     car['taxes'],
                                     car['consumption'],
                                     car['fuel_price'],
                                     car['repairs'],
                                     car['millage'],
                                     car['holding'],
                                     label=label,
                                     interest_rate=car['interest_rate'],
                                     rest_value=car['rest_value'])
    
    input_values_path = 'images/' + slugify(label) + '-boundaries.png'
    calc_df_path = 'images/' + slugify(label) + '-calc_df.png'
    
    latex = r"""%s
\twocolumn
\null
\vfill 
\includegraphics[width=\columnwidth]{%s}
\vfill 

\pagebreak

\null
\vfill 
\includegraphics[width=\columnwidth]{%s}
\vfill 

\pagebreak

\onecolumn
\null
\vfill 
\begin{center}
\includegraphics[width=0.6\columnwidth]{%s}
\end{center}
\vfill 
    """ %(latex,
          input_values_path,
          car['image_url'],
          calc_df_path
         )

    overview[label] = calc_values['Monatlich']
    dfi.export(input_values.round(2), input_values_path)
    dfi.export(calc_values.round(2), calc_df_path)
```
