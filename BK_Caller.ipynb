```python
latex = ''
for car in cars:
    label=car['label']
    (input_values, calc_values) = bk(car['car_price'],
                                     car['insurance'],
                                     car['taxes'],
                                     car['consumption'],
                                     car['fuel_price'],
                                     car['repairs'],
                                     car['millage'],
                                     car['holding'],
                                     label=label,
                                     interest_rate=car['interest_rate'],
                                     rest_value=car['rest_value'])
    
    input_values_path = 'images/' + slugify(label) + '-boundaries.png'
    calc_df_path = 'images/' + slugify(label) + '-calc_df.png'
    
    latex = r"""%s
\twocolumn

\section*{%s}
\begin{center}
\includegraphics[width=0.9\columnwidth]{%s}
\null
\vspace{0.5cm}
\includegraphics[width=0.9\columnwidth]{%s}
\end{center}

\pagebreak
\null
\vspace{2cm}
\begin{center}
\includegraphics[width=0.9\columnwidth]{%s}

%s, %d, Laufzeit: %.0f Jahre, \href{%s}{Weblink}
\end{center}

\pagebreak

""" %(latex,
      label,
      input_values_path,
      calc_df_path,
      car['image_url'],
      label,
      car['car_year'],
      car['holding'],
      car['web_url'].replace('#', '\#')
     )

    overview[label] = calc_values['Monatlich']
    
    
    sum_row = pd.IndexSlice[calc_values.index[calc_values.index == "SUMME"], :]
    dfi.export(input_values.round(2), input_values_path)
    dfi.export(calc_values.style.applymap(df_style, subset=sum_row).set_precision(2), calc_df_path)
```
